### **loss/perda e acurácia dos dados de treino e validação a cada Epoch**


````python
def loss_and_accuracy(history_passado):
  fig, ax = plt.subplots(ncols=2,figsize=(12,4))                                # Criando 2 gráficos, tamanho total: 12,4
  ax[0].plot(history_passado.history['loss'],label='loss')                      # Plotando a perda do treino
  ax[0].plot(history_passado.history['val_loss'], label='val_loss')             # Plotando a perda da validação
  ax[0].legend()                                                                # Definindo a legenda, dado nos plots anteriores
  ax[1].plot(history_passado.history['sparse_categorical_accuracy'],label='sparse_categorical_accuracy')               # Plotando a acurácia do treino
  ax[1].plot(history_passado.history['val_sparse_categorical_accuracy'], label='val_sparse_categorical_accuracy')      # Plotando a acurácia da validação
  ax[1].legend()                                                                # Definindo a legenda, dado nos plots anteriores
  plt.show()
````

### **Matrix de confusão com a quantidade de amostras também dada pelo usuário**


````python
def visualizar_matrix_confusao(x_teste, y_teste, modelo, amostra):
  prev = []                                                                     # Criando lista vazia paa armazenar as previsões
  for i in range(amostra):                                                      # Laço for onde o range é do tamanho passado na função
    pr = modelo.predict(np.expand_dims(x_teste[i], axis=0), verbose=0)          # Armazenando o predict de x_test[i] na variável "pr"
    prev.append(pr.argmax())                                                    # Armazenando isso na lista "prev"
  prev = np.array(prev)                                                         # Após o final do for, transformando a lista "prev" em array
  amostra = y_teste[:amostra]                                                   # Pegando a quantidade de valores de y_teste. Para comparamos os resultados
  return confusion_matrix(amostra, prev)                                        # Por fim, retorna a matrix de confusão.
````

### Gráfico com 20 primeiros dados de treino e a previsão da classe.

````python
def grafico_acertos_erros_modelo(modelo_passado, x_teste, y_teste):
  classes = ['Camisetas/top', 'Calças', 'Suéteres', 'Vestidos', 'Casacos',      # Obs: Apenas ao analisar a base que vi que essas são as classes. Depois disso que optei por armazená-las na variável "classes", e assim usar nessa função
            'Sandálias', 'Camisas', 'Tênis', 'Mochilas', 'Botas']
  fig, axs = plt.subplots(4,5, figsize=(10,10))                                 # Criando subplots com 4 linhas e 5 colunas, tamanho 10,10
  for i,ax in enumerate(axs.flat):                                              # Laço for para percorrer os axs. O flat é para deixar isso em uma dimensão
      ax.imshow(x_teste[i],cmap='gray')                                         # Exibindo a imagem no x_teste[i], onde o cmap é "gray"
      ax.axis('off')                                                            # Retirando os axis, para deixar a visualização mais limpa
      pred = modelo_passado.predict(np.expand_dims(x_teste[i], axis=0), verbose=0)             # Armazenando o predict na variável "pred"
      num_classe = y_teste[i]                                                                  # Armazenando o número da classe na variável "num_classe"
      ax.set_title(f'Item {y_teste[i]} - {classes[num_classe]}',pad=18, fontsize=10.5)         # Definindo o título, exibindo também a classe
      ax.text(14,-3,f'Previsão: {pred.argmax()} - {pred.max():.2%}', va='center', ha='center', # Exibindo a previsão com % e cor verde/vermelho para caso de acerto/erro
              color='green' if pred.argmax() == y_teste[i] else 'red')
  plt.show()
````

# **Criando o primeiro modelo de Redes Neurais - MLP**
- Obs: A explicações do passo a passo da criação do modelo estão nesse módulo
### **Características:**
- Modelo **simples**.
- Rescaling feito fora do modelo.
- Apenas **uma camada oculta**
- **Sem camadas de pré processamento.**


**Importando a base e definindo o treino e teste:**

````python
fashion_mnist = keras.datasets.fashion_mnist                                    # Importando o dataset
(xtrain, ytrain), (xtest, ytest) = fashion_mnist.load_data()                    # Separando os dados em treino e teste
````

**Rescaling:**

````python
xtrain = xtrain / 255.0                                                         # Alterando a escala dos dados de x de treino para valores entre 0 e 1
xtest = xtest / 255.0                                                           # Alterando a escala dos dados de x de teste para valores entre 0 e 1
````

**Criação do modelo:**

````python
modelo1 = keras.models.Sequential()
modelo1.add(keras.layers.InputLayer(input_shape=(28,28)))                       # Camada de entrada
modelo1.add(keras.layers.Dense(256, activation='relu'))                         # Camada oculta
modelo1.add(keras.layers.Flatten())                                             # Alterando a dimensão
modelo1.add(keras.layers.Dense(10, activation='softmax'))                       # Camada de saída
modelo1.summary()
````

**Definindo otimizador, função de custo, métrica e compilando o modelo:**

````python
otimizador = keras.optimizers.Adam()
funcao_custo = keras.losses.SparseCategoricalCrossentropy()
metrica = keras.metrics.SparseCategoricalAccuracy()

modelo1.compile(optimizer=otimizador, loss=funcao_custo, metrics=[metrica])
````

- **Criando o EarlyStop, definindo critérios como "patience":**
- **Fazendo o treinamento do modelo**

````python
parada_antecipada = keras.callbacks.EarlyStopping(patience=5, verbose=1, min_delta=1E-4)

history = modelo1.fit(xtrain, ytrain, epochs=200, batch_size=256, validation_split=0.1, callbacks=[parada_antecipada])
````

**Evaluate do modelo:**

````python
modelo1.evaluate(xtest, ytest)
````

**Visualizando a perda e acurácia nos dados de treino e validação:**

````python
loss_and_accuracy(history)
````

Analisando o gráfico acima, vemos um caso clássico de "overfitting", ou seja, o gráfico basicamente "decorou" os dados, tendo assim, uma má generalização.

- Obs: Atenção também na quantidade de ruídos e variações nas linhas dos dados de teste

- Obs: Função criada no módulo "Funções".

# **Criando o segundo modelo de Redes Neurais - CNN**
### **Características:**
- Modelo **mediano**
- **Rescaling** feito **dentro** do modelo
- Rede Neural **CNN**
- **Mais camadas ocultas**
- **Sem camadas** de processamento

````python
fashion_mnist = keras.datasets.fashion_mnist                                    # Importando dataset
(xtrain, ytrain), (xtest, ytest) = fashion_mnist.load_data()                    # Separando treino e teste
````

````python
modelo2 = keras.models.Sequential()                                             # Iniciando
modelo2.add(keras.layers.InputLayer(input_shape=(28,28,1)))                     # Definindo o formato da entrada
modelo2.add(keras.layers.Rescaling(scale=1./255))                               # Alterando a escala
modelo2.add(keras.layers.Conv2D(64,(3,3), activation='relu'))                   # Rede convolucional com 64 filtros, kernel de tamanho (3,3)
modelo2.add(keras.layers.Conv2D(64,(3,3), activation='relu'))                   # Rede convolucional com 64 filtros, kernel de tamanho (3,3)
modelo2.add(keras.layers.Conv2D(64,(3,3), activation='relu'))                   # Rede convolucional com 64 filtros, kernel de tamanho (3,3)
modelo2.add(keras.layers.Flatten())                                             # Mudando a dimensão
modelo2.add(keras.layers.Dense(10, activation='softmax'))                       # Camada de saída
modelo2.summary()
````

````python
optimizer = keras.optimizers.Adam(learning_rate=1E-3)                           # Otimizador: Adam
loss = keras.losses.SparseCategoricalCrossentropy()                             # loss: Crossentropy
metric = keras.metrics.SparseCategoricalAccuracy()                              # Metric: Accuracy

modelo2.compile(optimizer=optimizer, loss=loss, metrics=[metric])               # Compilando o modelo
````

````python
early_stop = keras.callbacks.EarlyStopping(patience=5, verbose=1, min_delta=1E-4 )                                  # Definindo a Parada Antecipada
history2 = modelo2.fit(xtrain, ytrain, epochs=200, batch_size=256, validation_split=0.2, callbacks=[early_stop])    # Fit do modelo
````

````python
modelo2.evaluate(xtest, ytest)
````

Tivemos um caso de **overfitting**, como podemos ver no gráfico abaixo:

````python
loss_and_accuracy(history2)
````

````python
grafico_acertos_erros_modelo(modelo2, xtest, ytest)
````

````python
visualizar_matrix_confusao(xtest,ytest,modelo2,1000)
````

# **Criando o terceiro modelo de Redes Neurais - CNN**
### **Características:**
- Modelo **melhorado**
- **Rescaling dentro** do modelo
- **Data Augmentation**
- **Dropout**

````python
fashion_mnist = keras.datasets.fashion_mnist                                    # Importando o dataset
(xtrain, ytrain), (xtest, ytest) = fashion_mnist.load_data()                    # Separando o treino e teste
````

````python
modelo3 = keras.models.Sequential()                                             # Iniciando
modelo3.add(keras.layers.InputLayer(input_shape=(28,28,1)))                     # Definindo o formato da entrada
modelo3.add(keras.layers.Rescaling(scale=1./255))                               # Alterando a escala

modelo3.add(keras.layers.RandomTranslation(height_factor=0.1, width_factor=0.1))# Data Augmentation: Transição da imagem
modelo3.add(keras.layers.RandomZoom(0.1))                                       # Data Augmentation: Zoom

modelo3.add(keras.layers.Conv2D(256,(3,3), activation='relu'))                  # Rede convolucional com 256 filtros, kernel de tamanho 3,3 e função de ativação ReLU
modelo3.add(keras.layers.Conv2D(256,(3,3), activation='relu'))                  # Rede convolucional com 256 filtros, kernel de tamanho 3,3 e função de ativação ReLU
modelo3.add(keras.layers.MaxPooling2D())                                        # MaxPooling, para resumir valores
modelo3.add(keras.layers.Dropout(0.2))                                          # Etapa de regularização, para evitar overfitting
modelo3.add(keras.layers.Conv2D(512,(3,3), activation='relu'))                  # Rede convolucional com 512 filtros, kernel de tamanho 3,3 e função de ativação ReLU
modelo3.add(keras.layers.MaxPooling2D())                                        # MaxPooling, para resumir valores
modelo3.add(keras.layers.Dropout(0.2))                                          # Etapa de regularização, para evitar overfitting

modelo3.add(keras.layers.Flatten())                                             # Alterando a dimensão
modelo3.add(keras.layers.Dense(10, activation='softmax'))                       # Camada de saída
modelo3.summary()
````

````python
optimizer = keras.optimizers.Adam(learning_rate=1E-3)                           # Otimizador: Adam
loss = keras.losses.SparseCategoricalCrossentropy()                             # Loss: CrossEntropy
metric = keras.metrics.SparseCategoricalAccuracy()                              # Metric: Accuracy

modelo3.compile(optimizer=optimizer, loss=loss, metrics=[metric])               # Compilando o modelo
````

````python
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss',patience=5, verbose=1, restore_best_weights=True ) # Parada antecipada monitoarando val_loss, com paciencia de 5 resultados. Ao ocorrer o early_stop, retorna os melhores pess
history3 = modelo3.fit(xtrain, ytrain, epochs=50, batch_size=512, validation_split=0.10, callbacks=[early_stop]) # Fit do modelo
````
